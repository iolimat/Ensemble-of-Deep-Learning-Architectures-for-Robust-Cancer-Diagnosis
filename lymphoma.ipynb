## Imports

```python
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.losses import BinaryCrossentropy
from tensorflow.keras import metrics
from keras import backend as K
```

## Helper functions

```python
def display_image_grid(dataset, num_images=8, grid_cols=4):
    """
    Display a grid of images from a TensorFlow dataset.

    Args:
        dataset (tf.data.Dataset): The TensorFlow dataset containing images.
        num_images (int, optional): Number of images to display. Default is 8.
        grid_cols (int, optional): Number of columns in the image grid. Default is 4.

    Returns:
        None: Displays a grid of images using matplotlib.

    This function takes a TensorFlow dataset containing images and displays a grid of specified
    number of images. It uses matplotlib to create the grid layout and display the images.

    Note:
        The input dataset is expected to contain tuples of (image, label). This function only
        displays the images.
    """

    dataset_iter = iter(dataset)
    num_rows = num_images // grid_cols  # Ceiling division
    fig, axes = plt.subplots(num_rows, grid_cols, figsize=(10, 10))


    for i in range(num_images):
        row = i // grid_cols
        col = i % grid_cols
        image = next(dataset_iter)[0][0]
        #print(image)
        image *= 255.0
        #print(image)
        axes[row, col].imshow(image.numpy().astype("uint8"))
        axes[row, col].axis("off")

    plt.tight_layout()
    plt.show()
```

```python
def rescale_image(image, label):
    '''
    Rescale the pixel values of an image to the range [0, 1].
    Parameters:
        image (tf.Tensor): The input image as a TensorFlow Tensor.
        label (tf.Tensor): The label corresponding to the image.
    Returns:
        tf.Tensor: The rescaled image with pixel values in the range [0, 1]
        tf.Tensor: The original label.

    Note:
      The input dataset is expected to contain tuples of (image, label).
    '''
    rescale_factor = 1./ 255.0
    image = tf.cast(image, tf.float32)

    image = image * rescale_factor
    return image, label
```

```python
def dispaly_reult(history):
    """
    Display training and validation loss, as well as training and validation accuracy, over epochs.

    Parameters:
        history (tf.keras.callbacks.History): The training history object obtained from model.fit().

    Returns:
        None: Displays the loss and accuracy curves using matplotlib.
    """

    h = history

    # Create a single figure with two subplots (1 row, 2 columns)
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 6))


    # Plot the loss value
    ax1.plot(h.history['loss'], label='Train loss')
    ax1.plot(h.history['val_loss'], label='Validation loss')
    ax1.legend()
    ax1.set_xlabel('Epochs', fontsize = 12)
    ax1.set_ylabel('Loss', fontsize = 12)
    ax1.set_title('Training and Validation Loss', fontsize = 15)



    # Plot the accuracy value
    ax2.plot(h.history['accuracy'], label='Train accuracy')
    ax2.plot(h.history['val_accuracy'], label='Validation accuracy')
    ax2.legend()
    ax2.set_xlabel('Epochs', fontsize = 12)
    ax2.set_ylabel('Accuracy', fontsize = 12)
    ax2.set_title('Training and Validation Accuracy', fontsize = 15)

    # Display the plots
    plt.show()
```

```python
def display_confusion_matrix(model, dataset,num_classes):
    """
    Display a confusion matrix heatmap for a given model's predictions on a dataset.

    Parameters:
      model (tf.keras.Model): The trained TensorFlow model.
      dataset (tf.data.Dataset): The dataset containing input images and corresponding labels.

    Returns:
      None: Displays the confusion matrix heatmap using matplotlib and seaborn.
    """

    y_true = np.concatenate([label.numpy() for _, label in dataset], axis=0)

    y_pred = model.predict(dataset.map(lambda image, label: image))
    if num_classes == 2:
        y_pred_labels = [1 if y >= 0.5 else 0 for y in y_pred]
    else:
        y_pred_labels = np.argmax(y_pred,axis = 1)
    cm = tf.math.confusion_matrix(y_true, y_pred_labels)
    cm = cm.numpy()/len(y_true)

    plt.figure(figsize=(14, 6))
    sns.heatmap(cm, annot=True, cmap='Greens',xticklabels = ticklabels, yticklabels = ticklabels)
    plt.title('Confusion Matrix', fontsize=20)
    plt.xlabel('Predicted', fontsize=15)
    plt.ylabel('Actual', fontsize=15)
    plt.show()
```

```python
def set_hyperparameter(path):
    """
    Set hyperparameters and paths based on the provided dataset directory structure.

    Args:
      path (str): The root directory path containing train, test, and val subdirectories.

    Returns:
      tuple: A tuple containing train_directory, test_directory, val_directory, loss,
              label_mode, and num_classes.

    This function takes the root directory path containing subdirectories for training, testing,
    and validation data. It determines appropriate hyperparameters based on the number of classes
    and returns them along with the directory paths.
    """

    train_directory = os.path.join(path, "train")
    test_directory = os.path.join(path, "test")
    val_directory = os.path.join(path, "val")
    loss = 'binary_crossentropy'
    label_mode = 'binary'
    num_classes = len(os.listdir(train_directory))
    if num_classes > 2:
        loss = 'sparse_categorical_crossentropy'
        label_mode = 'int'
    return train_directory, test_directory, val_directory, loss, label_mode, num_classes
```

## Hyperparameter

```python
path = '/kaggle/input/multi-cancer/Dataset/Lymphoma'
train_directory, test_directory, val_directory, loss, label_mode, num_classes = set_hyperparameter(path)

BATCH_SIZE = 64
EPOCHS = 50

image_size = (128, 128)
input_shape = (128, 128, 3)

patience = 5
monitor = 'val_loss'

ticklabels = os.listdir(test_directory)

cancer_name = 'Lymphoma'

target_size = (128, 128)
rotation_range = 20
width_shift_range = 0.1
height_shift_range = 0.1
shear_range = 0.2
rescale = 1./255
```

## Prepare Dataset

```python
train_dataset = tf.keras.utils.image_dataset_from_directory(
           directory = train_directory,
           label_mode = label_mode,
           batch_size = BATCH_SIZE,
           image_size = image_size,
           shuffle = True)
train_dataset = train_dataset.map(rescale_image)
```

```python
test_dataset = tf.keras.utils.image_dataset_from_directory(
           directory = test_directory,
           label_mode = label_mode,
           batch_size = BATCH_SIZE,
           image_size = image_size,
           shuffle = False)

test_dataset = test_dataset.map(rescale_image)
```

```python
val_dataset = tf.keras.utils.image_dataset_from_directory(
           directory = val_directory,
           label_mode = label_mode,
           batch_size = BATCH_SIZE,
           image_size = image_size,
           shuffle = False)

val_dataset = val_dataset.map(rescale_image)
```

## EDA

```python
display_image_grid(train_dataset.take(8), num_images=8, grid_cols=4)
```

## Training

```python
model = keras.Sequential([

    # input (128, 128, 3)
    keras.layers.InputLayer(input_shape = input_shape, batch_size = BATCH_SIZE),
    
    keras.layers.Conv2D(32, kernel_size=(3,3), padding='same', activation='relu'),
    keras.layers.BatchNormalization(),
    keras.layers.MaxPool2D(pool_size=(2,2)),
    # output (32, 32, 32)

    keras.layers.Conv2D(64, kernel_size=(3,3),  padding='same', activation='relu'),
    keras.layers.BatchNormalization(),
    keras.layers.MaxPool2D(pool_size=(2,2)),
    # output (16, 16, 64)

    keras.layers.Flatten(),
    # output (16*16*64)

    keras.layers.Dense(2048, activation='relu'),
    keras.layers.BatchNormalization(),
    
    keras.layers.Dense(512, activation='relu'),
    keras.layers.BatchNormalization(),
])

if num_classes == 2:    
    model.add(keras.layers.Dense(1, activation='sigmoid'))
else :
    model.add(keras.layers.Dense(num_classes, activation='softmax'))
```

```python
learning_rate = 0.0002

lr_scheduler = tf.keras.callbacks.LearningRateScheduler(
    lambda epoch: learning_rate * tf.math.exp(-epoch / 10), verbose=1
)

early_stopping = EarlyStopping(monitor = monitor, patience = patience, restore_best_weights = True)

optimizer = Adam()
model.compile(
    optimizer = optimizer,
    loss = loss,   
    metrics = ['accuracy'],
)
```

```python
history = model.fit(train_dataset,
                    epochs = EPOCHS,
                    validation_data = val_dataset,
                    callbacks = [lr_scheduler,early_stopping],
                   )
```

```python
dispaly_reult(history)
```

```python
model.evaluate(test_dataset)
```

```python
display_confusion_matrix(model, test_dataset,num_classes)
```
