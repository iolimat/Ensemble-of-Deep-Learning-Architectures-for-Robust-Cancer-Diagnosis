## Clean Dataset
Using the library cleanvision we can find all the issues in our dataset.

```python
def delete_files(file_paths):
    """
    Delete files at the specified file paths.

    Parameters:
    - file_paths (list of str): A list of file paths to be deleted.

    This function attempts to delete each file in the provided list of file paths.
    It prints a message indicating whether each file was successfully deleted or if there was an error.

    Returns:
    None
    """
    for file_path in file_paths:
        try:
            os.remove(file_path)
            print(f"File '{file_path}' has been deleted.")
        except OSError as e:
            print(f"Error deleting file '{file_path}': {e}")
```

```python
imagelab = Imagelab(data_path="/content/data")
imagelab.find_issues()
```

```python
# 6469 duplicates images!!

imagelab.issue_summary
```

```python
df = imagelab.issues
df.head(4).T
```

```python
file_paths = list(df.loc[(df["is_near_duplicates_issue"] == True) | (df["is_exact_duplicates_issue"] == True)].index)
```

```python
delete_files(file_paths)
```
